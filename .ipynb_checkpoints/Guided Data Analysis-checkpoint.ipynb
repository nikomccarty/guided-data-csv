## Pulling All Entries with a Certain Author

I am using the instructions given in the [Pandas Documentation](https://pandas.pydata.org/docs/user_guide/indexing.html). It is straightforward to pull out all rows in the Dataframe that contain a given value.

```python
# Let's pull out all rows that are for Elsevier as the Publisher.
df_elsevier = df[df['Publisher'] == 'Elsevier']
```

```python
len(df_elsevier)
```

```python
df_elsevier.head()
```

Of course, a _similar_ approach can be used to filter out even PARTIAL strings, such as a last name within the 'Author' column. For this part of the tutorial, I am following the Pandas documentation for [str.contains](https://pandas.pydata.org/docs/reference/api/pandas.Series.str.contains.html).

```python
#typing case=True means that it will filter in a case-sensitive manner. It will look precisely for 'Pamenter'
#setting na=False means that Python will pass by blank rows and fill them in with 'Nan'. If I do not set na=False, this command returns an error because there are blank values in your Dataframe.
df_author = df[df['Author'].str.contains('Pamenter', case=True, na=False)]
```

```python
df_author.head()
```

Let's do this for another author, just to ensure it is working!

```python
df_author2 = df[df['Author'].str.contains('Atangana', case=True, na=False)]
```

```python
len(df_author2)
```

You can use precisely this same technique to filter for a specific affiliation.

```python
df_institution = df[df['Institution'].str.contains('Zhejiang Cancer Hospital', case=True, na=False)]
```

```python
df_institution.head()
```

## Filter Retractions Based on 'Reasons'

Perhaps you want to pull out all rows that were retracted for this reason, but NOT that other reason. We can do this using a similar method, but now we are combining Boolean logic. For more details on this, see [this link](https://datascienceparichay.com/article/pandas-filter-dataframe-for-multiple-conditions/).

There are a few Boolean vectors available:

| means OR

& means AND

~ means NOT

These booleans can be combined using parentheses, as in this example.

```python
df_reasons = df[df['Reason'].str.contains('Error in Data', case=True, na=False) & df['Reason'].str.contains('Error in Text', case=True, na=False)]
```

```python
len(df_reasons)
```

```python
df_reasons.head()
```

```python
# To filter using NOT, we just add the ~ sign in front of the logic that we want to NEGATE. So here, we say DO NOT return a row if the reason is 'Error in Text'
df_reasons2 = df[df['Reason'].str.contains('Error in Data', case=True, na=False) & ~df['Reason'].str.contains('Error in Text', case=True, na=False)]
```

```python
df_reasons2.head()
```

## Pull out all authors, sort them alphabetically

```python
# First, let's pull out a single column from the DataFrame ... the column for authors.
authors = df['Author']
```

```python
# Now, the variable 'authors' consists of a single column, complete with all the authors.
authors
```

We can now go through each row, split the authors out by the ; value, and then sort them alphabetically.

```python
# We can iterate through each row, and split up the authors whenever we see a ';'
authors_split = authors.str.split(pat=';')
```

```python
authors_split
```

Now, let's remove all rows that are blank or empty. This will 'clean up' our data.

```python
authors_split = authors_split.dropna()
```
